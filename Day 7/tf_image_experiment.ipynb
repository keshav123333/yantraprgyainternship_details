```python
image_batch = tf.expand_dims(input=image, axis=0)
image_batch
```

```python
separate_filters_of_images_batch = tf.split(image_batch, num_or_size_splits=3, axis=-1)
```

```python
first_filter_of_image_batch = separate_filters_of_images_batch[0]
first_filter_of_image_batch
```

```python
sobel_filter = tf.constant([[-1, 0, 1], [-1, 0, 1], [-1, 0, 1]], dtype=tf.float32)
sobel_filter = tf.expand_dims(input=sobel_filter, axis=-1)
sobel_filter = tf.expand_dims(input=sobel_filter, axis=-1)
```

```python
sobel_filter
```

```python
first_filter_of_image_batch = tf.cast(x=first_filter_of_image_batch, dtype=tf.float32)
first_filter_of_image_batch
```

```python
tf.nn.conv2d
```

```python
convolved_image = tf.nn.conv2d(
    input=first_filter_of_image_batch, filters=sobel_filter, padding="SAME",
    strides=[1, 1, 1, 1],
)
convolved_image = tf.nn.relu(convolved_image)
```

```python
convolved_image
```

```python
for image in convolved_image:
    plt.imshow(image, cmap="gray")
    plt.show()
```
