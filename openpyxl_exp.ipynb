```python
def access_cells(ws):
    rows = []
    for i in range(1, ws.max_row+1):
        row = []
        print(f'cols:{ ws.max_column}')
        for j in range(1, ws.max_column+1):
            cell_value = ws.cell(row = i, column=j).value
            row.append(cell_value)
        print(len(row))
        rows.append(row)
    return rows
```

```python
list_of_lists_containg_excel_data = access_cells(ws = ws)
```

```python
list_of_lists_containg_excel_data
```
